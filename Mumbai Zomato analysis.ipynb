# Q10) Find the list of Affordable Restaurants?

#  The criteria to find the list of affordable restaurants  would be :
1) Low price 2) High Rating 

First step will be finding the restaurants with average cost 1/4th cost of most expensive restaurants in our dataframe

Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick the
restaurants that are 1/4th of 6000.


```python
max_price=zomato_mumbai['PRICE'].max()
one_fourth=max_price/4
one_fourth
```

```python
#Finding the cheapest restaurants which are having less than or equal to one fourth of the max price
aff_rest_df=zomato_mumbai[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE']]
aff_rest_df=aff_rest_df[aff_rest_df['PRICE']<=one_fourth]
aff_rest_df.sort_values(by='PRICE',inplace=True)
aff_rest_df
```

```python
#Finding the highly rated list of restuarants 
highrated_df=zomato_mumbai[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','RATING']]
highrated_df=highrated_df[highrated_df['RATING']>=4.5]
highrated_df.sort_values(by='PRICE',inplace=True)
highrated_df
```

# Now, we'll merge the aff_rest_df with highrate_rest_df to obtain the intersection i.e the list of Affordable Restaurants !!


```python
highrate_aff_df=pd.merge(aff_rest_df,highrated_df,how='inner',on=['NAME','REGION'])
highrate_aff_df=highrate_aff_df[['NAME','PRICE_x','CUSINE_CATEGORY_x','REGION','CUSINE TYPE_x']]
highrate_aff_df.rename(columns={'NAME':'NAME','PRICE_x':'PRICE','CUSINE_CATEGORY_x':'CUSINE_CATEGORY','REGION':'REGION',
                               'CUSINE TYPE_x':'CUSINE TYPE'},inplace=True)
highrate_aff_df
```

```python
# Affordable Restaurants with low price and high rating 

highrate_aff_df
```

# Q10) Find the list of most Reliable Restaurants?

# The Criteria for more reliable restaurants be
1)Low Price 2) High Rating 3) large number of votes
First step will be to find the restaurants with Votes greater than Mean of Votes

```python
mean_votes=zomato_mumbai['VOTES'].mean()
mean_votes
```

```python
highvotes_rest_df=zomato_mumbai[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','VOTES']]
highvotes_rest_df=highvotes_rest_df[highvotes_rest_df['VOTES']>=mean_votes]
highvotes_rest_df.sort_values(by='VOTES',inplace=True)
highvotes_rest_df
```

# These are the most reliable, highest rated and affordable restaurants:-

We obtain this dataframe by simply taking the intersection of highrate_aff_df & mean_rest_df

This dataframe obtained below shows the restaurants whose:

Cost is below 1250

Rating is above 4.5

Votes are above 177

```python
reliable_rest=pd.merge(highvotes_rest_df,highrate_aff_df,how='inner',on=['NAME','REGION'])
reliable_rest=reliable_rest[['NAME','PRICE_x','CUSINE_CATEGORY_x','REGION','CUSINE TYPE_x','VOTES']]
reliable_rest.rename(columns={'NAME':'NAME','PRICE_x':'PRICE','CUSINE_CATEGORY_x':'CUSINE_CATEGORY','REGION':'REGION','CUSINE TYPE_x':'CUSINE TYPE','VOTES':'VOTES'},inplace=True)
```
